Este notebook analiza la generación de residuos sólidos usando variables, input, print, operaciones aritméticas y conversión de datos.

```python
print("Diagnóstico básico de residuos sólidos")

punto = input("Ingrese el punto de monitoreo: ")

organicos = float(input("Residuos orgánicos recolectados (kg): "))
plasticos = float(input("Residuos plásticos recolectados (kg): "))
papel = float(input("Residuos de papel o cartón recolectados (kg): "))

total = organicos + plasticos + papel

porcentaje_organicos = (organicos / total) * 100
porcentaje_plasticos = (plasticos / total) * 100
porcentaje_papel = (papel / total) * 100

print("Punto de monitoreo:", punto)
print("Total de residuos recolectados:", total, "kg")
print("Porcentaje de residuos orgánicos:", porcentaje_organicos, "%")
print("Porcentaje de residuos plásticos:", porcentaje_plasticos, "%")
print("Porcentaje de papel o cartón:", porcentaje_papel, "%")
```
